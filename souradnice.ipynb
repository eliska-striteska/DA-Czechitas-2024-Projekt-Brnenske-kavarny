```python
import pandas as pd
import geopy
import json
```

```python
data_cafes = pd.read_csv('maps_google_scraped_final_dataset_V6.csv')
data_cafes.columns
```

```python
loc_data_cafes = data_cafes[['placeId', 'title', 'location']]
loc_data_cafes

# null = loc_data_cafes[loc_data_cafes['location'].isna() == True]
```

```python
print(loc_data_cafes['location'].apply(type).unique())
```

```python
loc_data_cafes['location'] = loc_data_cafes['location'].apply(lambda x: json.loads(x.replace("'", '"')))
```

```python
locations = pd.json_normalize(loc_data_cafes['location'])
locations
```

```python
loc_data_cafes_normalized = pd.concat([loc_data_cafes.drop('location', axis=1), locations], axis=1)
loc_data_cafes_normalized = loc_data_cafes_normalized.set_index('placeId')
loc_data_cafes_normalized
```

```python
loc_data_cafes_normalized.to_csv('location_cafes.csv')
```
