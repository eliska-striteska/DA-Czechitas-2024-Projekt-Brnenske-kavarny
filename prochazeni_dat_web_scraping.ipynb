```python
import pandas as pd
import json
```

```python
cafes = pd.read_json('dataset_google-maps-extractor_Brno_kavarny_cleaned.json')
cafes
```

```python
categories = cafes.groupby('categoryName')['title'].count().to_dict()

with open('list_of_categories.json', mode='w', encoding='utf-8') as output_file:
    json.dump(categories, output_file, indent=4, ensure_ascii=False)
```
