```python
import pandas as pd
import json
import collections
```

Závěrečné propojení datasetů:

```python
with open('dataset_google-maps-extractor_Brno_kavarny_cleaned.json', encoding= 'utf-8') as input_file_1:
    cafes_main = json.load(input_file_1)

with open('diff_cafe.json', encoding= 'utf-8') as input_file_2:
    cafes_2 = json.load(input_file_2)

with open('diff_coffee_shop.json', encoding= 'utf-8') as input_file_3:
    cafes_3 = json.load(input_file_3)

with open('diff_espresso_bar.json', encoding= 'utf-8') as input_file_4:
    cafes_4 = json.load(input_file_4)
```

```python
cafes_count = len(cafes_main) + len(cafes_2) + len(cafes_3) + len(cafes_4) 
cafes_count
```

```python
final_dataset = cafes_main + cafes_2 + cafes_3 + cafes_4

total_cafes_count = len(final_dataset)
total_cafes_count
```

```python
with open('maps_google_scraped_final_dataset_V2.json', mode= 'w', encoding= 'utf-8') as output_file:
    json.dump(final_dataset, output_file, indent= 4, ensure_ascii= False)
```

```python
titles = dict()

for cafe in final_dataset:
    if cafe['title'] not in titles.keys():
        titles[cafe['title']] = 1
    else:
        titles[cafe['title']] += 1

titles
```

```python
for title, number in titles.items():
    if number > 1:
        print(title, number)
```
