```python
import pandas as pd
import json
```

```python
def filter_keys_in_list_of_cafes(data, keys):
    # filtruje klíče ve slovnících "cafe" obsažených v celkovém seznamu
    # parametry: data = stažená data, keys = seznam klíčů, které chceme zachovat
    filtered_data = []
    for cafe in data:
        filtered_cafe = {}
        for key, value in cafe.items():
            if key in keys:
                filtered_cafe[key] = value
        filtered_data.append(filtered_cafe)
    return filtered_data
```

```python
with open('dataset_google-maps-extractor_2024-04-15_Brno_kavarny.json', encoding= 'utf-8') as file:
    kavarny_not_cleaned = json.load(file)

kavarny_clenaed = filter_keys_in_list_of_cafes(kavarny_not_cleaned, ['searchString', 'title', 'categoryName', 'address', 'street', 'city', 
               'postalCode', 'location', 'totalScore', 'reviewsCount', 'reviewsDistribution', 'openingHours', 'url'])

coffee_shop = []

for cafe in kavarny_clenaed:
    if cafe['categoryName'] == 'Coffee shop':
        coffee_shop.append(cafe)

with open('category_coffee_shop_form_orininal_dataset.json', mode='w', encoding= 'utf-8') as output_file:
    json.dump(coffee_shop, output_file, indent= 4, ensure_ascii= False)
    
```
