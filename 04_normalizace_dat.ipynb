Pokus o normalizaci dat vnořených v JSON strukturách

*Nakonec jsme tento kód nevyužili a data z JSON struktur rozbalovali až v SQL (JSON formát se nám později narušil stažením informací přes Nominatim API a ze souboru jsme udělali CSV), proto není kompletní a normalizuje jen sloupce [openingHours] a [categories]*

```python
import pandas as pd
import json
```

```python
with open('maps_google_scraped_final_dataset_V2.json', mode='r', encoding= 'utf-8') as input_file:
    data = json.load(input_file)

data
```

```python
df_opening_hours = pd.json_normalize(data, 'openingHours', meta='placeId')
df_opening_hours
```

```python
df_categories = pd.json_normalize(data, 'categories', meta='placeId')
df_categories
```

```python
df_data = pd.read_json('maps_google_scraped_final_dataset_V2.json')
df_data
```

```python
df_merged = pd.merge(df_data, df_opening_hours, on='placeId')
df_merged = pd.merge(df_merged, df_categories, on='placeId')
df_merged
```

```python
print(df_merged.columns)
```

```python
df_final = df_merged.drop('openingHours', axis=1)
df_final = df_final.drop('categories', axis=1)
df_final.rename(columns={0: 'category_other'}, inplace=True)
df_final
```

```python
df_final.to_csv('maps_google_scraped_final_dataset_V3.csv')
```
